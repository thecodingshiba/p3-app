# PROJECT 3 - CLEAN DATA

```python
import pandas as pd
import os
```

## GDP_Data

```python
# Set up variable
folder_path="resources"
file_name="GDP_Data.csv"
file_path=os.path.join(folder_path,file_name)
results_path=os.path.join('results','cleaned_'+file_name)
```

```python
# Load data set
gdp_data=pd.read_csv(file_path, skiprows=2) # as the table starts from line 3
```

#### Explore the Data

```python
# First 5 row:
gdp_data.head()
```

```python
# Display data types and missing values
gdp_data.info()
```

```python
# Display summary statistics
gdp_data.describe()
```

```python
# Handle missing value
gdp_data=gdp_data.fillna(0)
```

```python
# Reshape data -> unpivot year from columns->rows
gdp_data_melted = pd.melt(gdp_data, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='GDP')
gdp_data_melted.head()
```

```python
# Correct Data Types
gdp_data_melted['Country Name'] = gdp_data_melted['Country Name'].astype(str)
gdp_data_melted['Country Code'] = gdp_data_melted['Country Code'].astype(str)
gdp_data_melted['Indicator Name'] = gdp_data_melted['Indicator Name'].astype(str)
gdp_data_melted['Indicator Code'] = gdp_data_melted['Indicator Code'].astype(str)
gdp_data_melted['Year'] = gdp_data_melted['Year'].astype(int)
gdp_data_melted.info()
```

```python
# Drop 2 columns['Indicator Name'],['Indicator Code']
gdp_data_melted = gdp_data_melted.drop(['Indicator Name', 'Indicator Code'], axis=1)
```

```python
# Save Cleaned Data
gdp_data_melted.to_csv(results_path,index=False)
```

```python
# Display data types and missing values
population_data.info()
```

```python
# Display summary statistics
population_data.describe()
```

```python
# Handle missing values
population_data=population_data.fillna(0)
```

```python
# Correct Data Types
population_data['Year'] = population_data['Year'].astype(int)
# Get the list of headers
headers = population_data.columns.tolist()
#Remove all space in the data
population_data = population_data.applymap(lambda x: x.strip() if isinstance(x, str) else x)
#Change type to float for all column from 5->58
for i in range(4,len(headers)):
    try:
        population_data[headers[i]] = population_data[headers[i]].replace('...', 0) #There was a wierd elipsis on the way >.<
        population_data[headers[i]] = population_data[headers[i]].replace('[^\d.]', '', regex=True) #Remove annoying spaces as thousands seperator
    except:
        pass
    population_data[headers[i]] = population_data[headers[i]].astype(float)
```

```python
# Check types
population_data.info()
```

```python
# As the regions do not have ISO3 and ISO2 -> get all that have NA value
region_population_data = population_data[(population_data['ISO3 Alpha-code'] == 0) | (population_data['ISO2 Alpha-code'] == 0)]
region_population_data = region_population_data.drop(['ISO2 Alpha-code','ISO3 Alpha-code'], axis=1).reset_index(drop=True) # drop unnescessary columns
# As the specific contries will have ISO3 and ISO2 -> drop all that have NA value
population_data = population_data[(population_data['ISO3 Alpha-code'] != 0) | (population_data['ISO2 Alpha-code'] != 0)]
```

```python
# Export the results to csv
population_data.to_csv(results_path,index=False)
region_population_data.to_csv(results_path.replace(file_name, 'region_population_data.csv'), index=False)
```
